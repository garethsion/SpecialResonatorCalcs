```python
import numpy as np
import csv

def read_1D_comsol_data(file):
    x=[]
    y=[]
    with open(file, 'r') as rf:
        reader = csv.reader(rf, delimiter=',')
        for row in reader:
            x.append(row[0])
            y.append(row[1])
    x = np.asarray((x),dtype=float)
    y = np.asarray((y),dtype=float)
    return x,y

def read_2D_comsol_data(file):
    x=[]
    y=[]
    z=[]
    with open(file, 'r') as rf:
        reader = csv.reader(rf, delimiter=',')
        for row in reader:
            x.append(row[0])
            y.append(row[1])
            z.append(row[2])
    x = np.asarray((x),dtype=float)
    y = np.asarray((y),dtype=float)
    z = np.asarray((z),dtype=float)
    return x,y,z
```

```python
xx,Bx = read_1D_comsol_data('Bx.csv')
xy,By = read_1D_comsol_data('By.csv')
xn,Bn = read_1D_comsol_data('normB.csv')

from matplotlib import pyplot as plt
# plt.figure(figsize=(6,4))
fig, (ax1, ax2) = plt.subplots(1,2,sharex=True,figsize=(12,8))
ax1.plot(xx,Bx,LineWidth=4,color='r',label='$\delta B_{x}$')
ax1.plot(xy,By,LineWidth=4,color='g',label='$\delta B_{y}$')
ax1.legend(fontsize=16)
fig.text(0.5,-0.04, '$x (\mu m)$', ha='center',fontsize='28')
# plt.xlabel('$x (\mu m)$',fontsize=28,ha='centre')
ax1.set_ylabel('$B (T)$',fontsize=28)
ax2.plot(xn,Bn,LineWidth=4,color='b',label='$B_{norm}$')
ax2.legend(fontsize=16)
plt.tight_layout()
plt.show()
```
